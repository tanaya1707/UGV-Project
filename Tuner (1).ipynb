```python
def build_model(hp):
    model = keras.Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(layers.Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=16),
                               activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='categorical_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
lst= []

for i in range(0, 10000):
   lst.append([image[i], Bacterial_spot[i], Early_blight[i], Late_blight[i], Leaf_Mold[i], Septoria_leaf_spot[i], Two_spotted_spider_mite[i], Target_Spot[i], Tomato_Yellow_Leaf_Curl_Virus[i], Tomato_mosaic_virus[i], healthy[i]])
df= pd.DataFrame(lst, columns= ['Image', 'Bacterial_spot', 'Early_blight', 'Late_blight', 'Leaf_Mold', 'Septoria_leaf_spot', 'Two_spotted_spider_mite', 'Target_Spot', 'Tomato_Yellow_Leaf_Curl_Virus', 'Tomato_mosaic_virus', 'healthy'])
df
```

```python
from sklearn.model_selection import train_test_split
X= df.iloc[:, 1]
Y= df.iloc[:, 1:len(lst)]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory(
    directory='/content/drive/MyDrive/UGV_AI/tomato/train',
    target_size=(256, 256),
    color_mode="rgb",
    batch_size=16,
    class_mode="binary",
    shuffle=True,
)
test_set = test_datagen.flow_from_directory(
    directory=r'/content/drive/MyDrive/UGV_AI/tomato/val',
    target_size=(256, 256),
    color_mode="rgb",
    batch_size=16,
    class_mode="binary",
    shuffle=True,
)
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory=r'C:\Users\Amit Gupte\Desktop\UGV Project\tomato',
    project_name='Keras_tuner')
```

```python
tuner.search(X_train, y_train, epochs=50, validation_data= (X_test, y_test))
```
